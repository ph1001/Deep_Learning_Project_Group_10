```python
import pandas as pd
import numpy as np
import plotly as py
import matplotlib.pyplot as plt
```

```python
base = pd.read_excel('01 Base.xlsx', index_col=0)
base_DO = pd.read_excel('02 Base+DO.xlsx', index_col=0)
base_DO_DA = pd.read_excel('03 Base+DO+DA.xlsx', index_col=0)
```

```python
base.head(3)
```

```python
base_DO.head(3)
```

```python
base_DO_DA.head(3)
```

### Plot accuracies

#### Create a DF to plot the accuracies (train and val)

```python
final_acc = pd.DataFrame()
```

```python
final_acc_columns = ['Train Acc Base', 'Train Acc Base + DO', 'Train Acc Base + DO + DA', 'Val Acc Base', 'Val Acc Base + DO', 'Val Acc Base + DO + DA']
```

```python
for column in final_acc_columns:
    final_acc[column] = None
```

```python
final_acc
```

```python
final_acc['Train Acc Base'] = base['acc']
final_acc['Train Acc Base + DO'] = base_DO['acc']
final_acc['Train Acc Base + DO + DA'] = base_DO_DA['acc']
final_acc['Val Acc Base'] = base['val_acc']
final_acc['Val Acc Base + DO'] = base_DO['val_acc']
final_acc['Val Acc Base + DO + DA'] = base_DO_DA['val_acc']
```

```python
final_acc
```

#### Plots

```python
# define area
plt.figure(figsize=(10, 5), dpi=80)

# define x axis
x = final_acc.index
  
# plot lines
plt.plot(x, final_acc['Train Acc Base'], label = "Train Acc Base", color='green')
plt.plot(x, final_acc['Train Acc Base + DO'], label = "Train Acc Base + DO", color='blue')
plt.plot(x, final_acc['Train Acc Base + DO + DA'], label = "Train Acc Base + DO + DA", color='red')
plt.plot(x, final_acc['Val Acc Base'], label = "Val Acc Base", color='green', linestyle='-.')
plt.plot(x, final_acc['Val Acc Base + DO'], label = "Val Acc Base + DO", color='blue', linestyle='-.')
plt.plot(x, final_acc['Val Acc Base + DO + DA'], label = "Val Acc Base + DO + DA", color='red', linestyle='-.')
plt.xticks(np.arange(0, 20, 1))
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Train and Val Accuracies")
plt.title("Comparing Base Model, Base Model + Dropout and Base Model + Dropout + Data Augmentation - Accuracy")
plt.show()
```
